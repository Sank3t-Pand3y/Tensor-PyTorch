```python
# Exmaple to practise nn.Module

import torch
import torch.nn as nn
```

```python
# create model class

class Model(nn.Module):  # The Model class that i created should be inherit from nn.Module.
                           # if we don't do this then we can access nn.Module
  def __init__(self, num_features):

    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features,3),
        nn.ReLU(),
        nn.Linear(3,1),
        nn.Sigmoid()
    )

  def forward(self, features):
    out = self.network(features)
    return out
```

```python
# Create a fake Dataset

features = torch.rand(10,5)
```

```python
# create a model

model = Model(features.shape[1])

# call Model for forward pass

model(features)
```
